```python
# 2d connectivity matrix
mat = np.array([[0,5,0],
                [0,0,20],
                [0,0,0]])
```

```python
# # len(net.nodes) # number of nodes
# # for (u, v, c) in net.edges.data
# # net.edges.data("weight")

# # convert  net.edges(data=True) to a dict
# # [(u, v, c) for u, v, c in net.edges.data('weight')]

# [(u, v, c) for u, v, c in net.edges(data=True)]
```

```python
# networkx object
net = nx.from_numpy_array(mat, create_using=nx.DiGraph)

N = len(net.nodes) # number of nodes
E = len(net.edges) # number of edges

# nodes
node_size = np.array([2,4,8]) *2
node_title = ['a','b','c']
node_level = [2,1,3]
# node_value = [1] #scaling
nx.set_node_attributes(net, dict(zip(np.arange(N), node_size)), 'size')  # dict(net.degree)
nx.set_node_attributes(net, dict(zip(np.arange(N), node_title)), 'title')
nx.set_node_attributes(net, dict(zip(np.arange(N), node_level)), 'level')
# nx.set_node_attributes(net, dict(zip(np.arange(N), node_value)), 'value')

# edges
# edge_width = [(u, v, c) for u, v, c in net.edges.data('weight')]
# nx.set_edge_attributes(net, edge_width, 'width')
```

```python
# pyvis object
net2 = Network(directed=True, layout='hierarchical', notebook=False,
               neighborhood_highlight=True, select_menu=False, filter_menu=True)
net2.height = 600 #"900px" "75%"
net2.width = 600
net2.from_nx(net, 
            node_size_transf=lambda x: x, 
            edge_weight_transf=lambda x: x # if keep weight as is, can use to scale edge width
            )

# net2.toggle_physics(False)
# net2.show_buttons(filter_=['node', 'edge', 'physics']) # [nodes, edges, layout, interaction, manipulation, physics, selection, renderer]
net2.show_buttons(filter_=['layout'])

# net2.set_options("""
# """)

net2.show("test_pyvis.html", notebook=False)
```
